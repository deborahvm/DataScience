```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8)
```

```python
data[50:100, 0] # sepalas das iris-setosas
```

```python
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8, label='Com. Spala Iris-Setosa')
plt.plot(data[50:100,0], c='Black', ls=':', marker='o', ms=8, label='Com. Spala Iris-Versicolor')
plt.legend()
plt.show()
```
