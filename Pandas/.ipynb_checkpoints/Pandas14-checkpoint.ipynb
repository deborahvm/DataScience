```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,100)
```

```python
plt.plot(x,np.sin(x), 'b--')
plt.plot(x,np.cos(x), 'r--')
plt.show()
```
